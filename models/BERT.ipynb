```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchtext.datasets import IMDB
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from tqdm import tqdm
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, AdamW

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
print("✅ Transformers working, tokenizer loaded successfully!")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
train_iter, test_iter = IMDB(split=('train', 'test'))
train_data = list(train_iter)
test_data = list(test_iter)

print("Train samples:", len(train_data))
print("Test samples:", len(test_data))
print("Example sample:", train_data[0])
```

```python
# Using 2000 reviews for training and 1000 for testing
import random
small_train = random.sample(train_data, 2000)
small_test = random.sample(test_data, 1000)
print(f"Using subset: {len(small_train)} train and {len(small_test)} test samples.")
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
MAX_LEN = 128  # reduced for faster training
```

```python
def encode_batch(batch):
    texts = [text for (label, text) in batch]
    labels = [1 if label == 'pos' else 0 for (label, text) in batch]
    encodings = tokenizer(
        texts,
        padding=True,
        truncation=True,
        max_length=MAX_LEN,
        return_tensors='pt'
    )
    encodings['labels'] = torch.tensor(labels)
    return encodings
```

```python
BATCH_SIZE = 8

train_dataloader = DataLoader(small_train, batch_size=BATCH_SIZE, shuffle=True, collate_fn=encode_batch)
test_dataloader = DataLoader(small_test, batch_size=BATCH_SIZE, shuffle=False, collate_fn=encode_batch)
```

```python
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
model.to(device)
```

```python
optimizer = AdamW(model.parameters(), lr=2e-5)
```

```python
EPOCHS = 3

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    correct, total = 0, 0

    for batch in tqdm(train_dataloader, desc=f"Epoch {epoch+1} Training"):
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        preds = torch.argmax(outputs.logits, dim=1)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    avg_loss = total_loss / len(train_dataloader)
    accuracy = 100 * correct / total
    print(f"Epoch {epoch+1} - Train Loss: {avg_loss:.4f}, Train Accuracy: {accuracy:.2f}%")

    # --- Evaluation after each epoch ---
    model.eval()
    correct, total, eval_loss = 0, 0, 0
    with torch.no_grad():
        for batch in test_dataloader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            preds = torch.argmax(outputs.logits, dim=1)

            eval_loss += loss.item()
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    print(f"Epoch {epoch+1} - Val Loss: {eval_loss/len(test_dataloader):.4f}, Val Accuracy: {100*correct/total:.2f}%\n")
```
